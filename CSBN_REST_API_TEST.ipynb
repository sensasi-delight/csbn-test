# Initialization
Import Required Package

```python
import requests
import json
import pandas
import copy

pandas.set_option('display.max_columns', None)
pandas.set_option('display.max_rows', None)
pandas.set_option('display.max_colwidth', None)
```

import dataset for test case

```python
DATASET_FOLDER_URL = 'https://raw.githubusercontent.com/sensasi-delight/csbn-test/main/dataset/'

PRODUCTS = requests.get(DATASET_FOLDER_URL + 'products.json').json()
BATCHES = requests.get(DATASET_FOLDER_URL + 'batches.json').json()
INVOICES = requests.get(DATASET_FOLDER_URL + 'invoices.json').json()
SHIPMENTS = requests.get(DATASET_FOLDER_URL + 'shipments.json').json()
```

Test Function Definition

```python
host_url = "http://128.199.185.208:3000"

column0 = []
column1 = []
column2 = []
column3 = []
column4 = []
column5 = []


def handler(channel, peer, function, data):
  column0.append(channel)
  column1.append(peer)
  column2.append(function)
  column3.append(data)

  if function[:6] == "create":
    url = host_url + "/" + function + "/" + channel + "/" + peer
    response = requests.post(url, json = data).json()
  elif function[:4] == "read":
    url = host_url + "/" + function + "/" + channel + "/" + peer + "/" + data
    response = requests.get(url).json()
  elif function[:6] == "update":
    url = host_url + "/" + function + "/" + channel + "/" + peer + "/" + data['id']
    response = requests.put(url, json = data).json()

  column4.append(response)
  column5.append(response['success'])
```
